```python
##QUESTION-1
import thinkdsp as t
import thinkplot as p

s1 = t.SinSignal(freq = 24, amp = 3, offset = 0)
s2 = t.SinSignal(freq = 5, amp = 3, offset = 0)
s3 = t.SinSignal(freq = 1999, amp = 2, offset = 0)

mix = s1+s2+s3
wave = mix.make_wave(start = 0, duration = 3, framerate = 48000)

wave.make_spectrum().plot()
```
